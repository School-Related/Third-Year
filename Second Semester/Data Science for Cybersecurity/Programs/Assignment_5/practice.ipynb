```python
# lets import sklearn things
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# import dataset
import seaborn as sns

df = sns.load_dataset("tips")
```

```python
df
```

```python
sns.pairplot(df)  # basic eda
```

```python
# lets do total bill and tip
sns.lmplot(df, x="tip", y="total_bill")
```

```python
# lets implement linear regression and just see if its kinda the same
model = LinearRegression()
```

```python
# split
xtr, xtst, ytr, ytst = train_test_split(
    df[["tip"]], df["total_bill"], train_size=0.8, random_state=42
)
xtr
```

```python
# fit the model
model.fit(xtr, ytr)
```

```python
# now we can predict
ypred = model.predict(xtst)
```

```python
# lets plot
sns.lineplot(x=xtst["tip"], y=ypred)
sns.scatterplot(x=xtst["tip"], y=ytst)
```

```python
sns_test_df = xtst
sns_test_df["y"] = ytst
sns_test_df
```

```python
sns.lmplot(sns_test_df, x="tip", y="y", ci=99)
```
