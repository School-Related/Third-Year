```python
import matplotlib.pyplot as plt
import pandas as pd

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

sns.scatterplot(x=x, y=y, hue=y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

output_file("scatter_plot.html")

p = figure(title="Scatter Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
p.circle(x, y, size=10, color="navy", alpha=0.5)

show(p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
y2 = [i + 1 for i in y]
x2 = [i + 1 for i in x]

data_dict = {"x": [], "y": [], "data": []}

for i in range(len(x)):
    data_dict["x"].append(x[i])
    data_dict["y"].append(y[i])
    data_dict["data"].append("A")

    data_dict["x"].append(x2[i])
    data_dict["y"].append(y2[i])
    data_dict["data"].append("B")
df = pd.DataFrame(data_dict)
sns.lineplot(x=df["x"], y=df["y"], hue=df["data"], style=df["data"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

output_file("line_plot.html")

p = figure(title="Line Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
p.line(x, y, line_width=2)

show(p)
```
